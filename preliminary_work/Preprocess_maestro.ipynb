```python
import numpy as np
import pypianoroll as ppr
import glob
import pickle

# Paths to the data
path_train = '/kaggle/input/maestro/train/*.mid'
path_valid = '/kaggle/input/maestro/validation/*.mid'
path_test = '/kaggle/input/maestro/test/*.mid'

path_iter = iter(sorted(glob.glob(path_train)))
```

```python
def midi2pr(filename, chunk_size = 128):
    """
    This functions receives a filename, reads the associated midi file,
    extracts the piano roll and returns a list of arrays of size (chunk_size,128)
    """
    pr = ppr.parse(filename, beat_resolution = 8)
    pr = pr.tracks[0].pianoroll
    time = pr.shape[0]
    a =  np.array([pr[i:i+128] for i in range(0, time-chunk_size, chunk_size)])
    return a
```

```python
path_iter = iter(sorted(glob.glob(path_train)))

tr_array = midi2pr(next(path_iter))

for filename in path_iter:
    tr_array = np.concatenate((tr_array, midi2pr(filename)))
```

```python
path_iter = iter(sorted(glob.glob(path_valid)))

val_array = midi2pr(next(path_iter))

for filename in path_iter:
    val_array = np.concatenate((val_array, midi2pr(filename)))
```

```python
path_iter = iter(sorted(glob.glob(path_test)))

test_array = midi2pr(next(path_iter))

for filename in path_iter:
    test_array = np.concatenate((test_array, midi2pr(filename)))
```

```python
pickle.dump(tr_array, open("train_128.pickle", 'wb'))
pickle.dump(val_array, open("valid_128.pickle", 'wb'))
pickle.dump(test_array, open("test_128.pickle", 'wb'))
```
